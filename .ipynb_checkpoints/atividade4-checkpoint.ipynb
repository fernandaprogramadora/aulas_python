## Exemplos

```python
#Verificação de maioridade
idade_usuario = int(input("Digite a sua idade:"))
idade_usuario
```

```python
if idade_usuario >=18:
    print("Você tem idade para entrar")
else:
     print("Você tem idade para entrar")
```

```python
#Verificar a aprovação ou não 
nota_aluno = int(input("Digite a sua média final"))
nota_aluno
```

```python
if nota_aluno >= 7:
    print("Parabéns, você foi aprovado (a)!")

elif nota_aluno >= 5:
    print("Você está em recuperação")

else:
     print("Que pena, você foi reprovado (a)!")
```

```python
#Verificar '==' e '!='
numero_digitado = float(input ("Digite um número entre -2 e 5"))
numero_digitado
```

```python
if numero_digitado == -1:
    print("Você acertou!")

else:
     print("Que pena, tente novamente!")
    
```

```python
if numero_digitado != 3:
    print("Que pena, tente novamente!")

else:
     print("Você acertou!")
```

```python
#Resto da divisão por um número '%'
#Par ou ímpar
numero = int(input("Digite um número natural")) 
if numero%2 == 0:
    print(f"O {numero} é par!")

else:
    print(f"O {numero} é ímpar!")
```

```python
#Comparar dois números
num1 = float(input("Digite o primeiro número:"))
num2 = float(input("Digite o segundo número:"))

if num1 > num2:
    print(f"{num1} > {num2}")
elif num1 == num2:
    print(f"{num1} = {num2}")
else:
    print(f"{num2} > {num1}")
```

```python
#Resolvendo uma equação do segundo grau
a = float(input("Digite o valor do coeficiente a"))
b = float(input("Digite o valor do coeficiente a"))
c = float(input("Digite o valor do coeficiente a"))

delta = b**2 - 4*a*c
x1 = (-b + delta**(0.5))/(2*a)
x2 = (-b - delta**(0.5))/(2*a)

if delta < 0:
    print(f"Delta menor do zero: {delta}")
    print("Não há raíz real")
elif a == 0:
    print(f" {a} é menor que zero, insira um número diferente de 0")
elif delta >= 0:
    print(f"x' = {x1} e x'' = {x2}")
else:
    print(f"x'={x1} e x'' = {x2}")
```
