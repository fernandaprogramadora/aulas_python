Crie um programa que:
1. Peça a idade do usuário.
2. Mostre na tela:
▶ Criança (se for menor de 13 anos)
▶ Adolescente (de 13 a 17 anos)
▶ Adulto (18 anos ou mais)

```python
idade = int(input("Qual a sua idade?"))
if (idade <= 12):
    print("Criança")
elif(idade <= 17):
    print("Adolescente")
else:
    print("Adulto")
```

5. Aprovação de aluno
Peça a nota do aluno. Se a nota for maior ou igual a 7,
mostre "Aluno aprovado". Caso contrário, mostre "Aluno
reprovado".

```python
nota = int(input("Nota do aluno:"))
if (nota >= 7):
    print("Aprovado")
else:
    print("Reprovado")
```

6. Comparar dois números
Peça dois números ao usuário. Mostre qual é maior.
Exemplo: "O maior número é 25".

```python
numero1 = int(input("Digite o primeiro número:"))
numero2 = int(input("Digite o segundo número:"))
if (numero1 > numero2 ):
    print("Qual é maior?")
else:
    print("O maior número é 25")
```
